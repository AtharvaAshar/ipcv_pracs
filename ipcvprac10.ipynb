Prewitts

```python
f1_p=np.array([[-1,-1,-1], [0,0,0], [1,1,1] ])
f2_p=np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
f_combine=f1_p +f2_p
```

```python
def convolution(img,filter1):
    output_p=[]
    for r in range(img.shape[0]-2): 
      temp = []
      for c in range(img.shape[1]-2):
        lhs = img[r:r+3, c:c+3]
        ans = 0 
        for r_f in range(filter1.shape[0]): 
          for c_f in range(filter1.shape[1]):
            ans += lhs[r_f][c_f]*filter1[r_f][c_f] 
        temp.append(ans) 
      output_p.append(temp)
    plt.imshow(output_p,cmap="gray")
```

```python
# first filter
convolution(img,f1_p)
```

```python
convolution(img,f2_p)
```

```python
convolution(img,f_combine)
```

Sobel

```python
f1_s=np.array([[-1,-2,-1], [0,0,0], [1,2,1] ])
f2_s=np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
f_combine_s=f1_p +f2_p
```

```python
# first filter
convolution(img,f1_s)
```

```python
# 2nd filter
convolution(img,f2_s)
```
