```python
import cv2
import matplotlib.pyplot as plt
import copy
import numpy as np
```

```python
img=cv2.imread("pout.png",0)
# img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap="gray")
```

```python
s_max,s_min,r_max,r_min=255,0,img.max(),img.min()
```

```python
def eqn(s_max,s_min,r_max,r_min,r):
  coeff=(s_max-s_min)/(r_max-r_min)
  constant= -(coeff*r_min) +s_min 
  v=coeff*r + constant
  v=int(round(v))
  return v
```

```python
result=img.flatten()
bins=[i for i in range(0,256)]
plt.hist(result,bins=bins)
plt.show()
```

```python
img_copy= copy.deepcopy(img)
old_val=img_copy[img_copy>= r_min]
set_old_val=list(set(old_val))
```

```python
set_new_val=[]
for i in set_old_val:
  set_new_val.append(eqn(s_max,s_min,r_max,r_min,i))
print(set_new_val)
```

```python
for i,j in zip(set_old_val,set_new_val):
  img_copy[img_copy==i]=j
```

```python
result_op=img_copy.flatten()
bins=[i for i in range(0,256)]
plt.hist(result_op,bins=bins)
plt.show()
```

```python
(unique, counts)= np.unique(img,return_counts=True)
frequencies=np.asarray((unique,counts)).T
```

```python
final_frequency_key=[]
final_frequency=[]
for i in frequencies:
  final_frequency_key.append(i[0])
  final_frequency.append(i[1])
```

```python
final_frequency
```

```python
for i in range(0,256):
  if i not in final_frequency_key:
    final_frequency_key.insert(i,i)
    final_frequency.insert(i,0)
print(final_frequency)
```

```python
pdf=[]
for i in final_frequency:
  pdf.append(round(i/sum(final_frequency),3))
print(pdf)
```

```python
cdf=[pdf[0]]
for i in range(1,len(pdf)):
  cdf.append(pdf[i]+cdf[-1])
print(cdf[-1])
```

```python
l_1_cdf=[]
for i in cdf:
  i=i*255
  l_1_cdf.append(int(round(i,0)))
print(l_1_cdf)
```

```python
mydct={}
for i in range(1,len(l_1_cdf)):
  if l_1_cdf[i]==l_1_cdf[i-1]:
    r+=final_frequency[i-1]
  else:
    r+=final_frequency[i-1]
    mydct[l_1_cdf[i-1]]=r 
    r=0

if l_1_cdf[-1] not in list(mydct.keys()):
  mydct[l_1_cdf[-1]]=final_frequency[-1]
print(mydct)
```

```python
plt.bar(mydct.keys(),mydct.values(),color="black")
plt.show()
```

```python
import copy
img_copy = copy.deepcopy(img) 
for i, j in zip(final_frequency_key, l_1_cdf): 
  img_copy[img_copy==i] = j
```
