```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Erosion

```python
img1= cv2.imread("e.png",0)
```

```python
m,n= img1.shape 
```

```python
plt.imshow(img1, cmap="gray")
```

```python
k=15
SE=np.ones((k,k),dtype=np.uint8)
constant=(k-1)//2
```

```python
output=np.zeros((m,n),dtype=np.uint8)
```

```python
for i in range(constant,m-constant):
  for j in range(constant,n-constant):
    temp=img1[i-constant:i+constant+1,j-constant:j+constant+1]
    product=temp*SE
    output[i][j]=np.min(product)
plt.imshow(output,cmap="gray")
```

Dialation

```python
img2=cv2.imread("text.png",0)
plt.imshow(img2,cmap="gray")
```

```python
p,q=img2.shape
output2=np.zeros((p,q))
```

```python
SED=[[0,1,0],[1,1,1],[0,1,0]]
```

```python
constant1=1
for i in range(constant1,p-constant1):
  for j in range(constant1,q-constant):
    temp=img2[i-constant1:i+constant1+1,j-constant1:j+constant1+1]
    product= temp * SED
    output2[i][j]= np.max(product)
plt.imshow(output2,cmap="gray")
```

```python
cv2.imwrite("Dilated.png", img2)
```
