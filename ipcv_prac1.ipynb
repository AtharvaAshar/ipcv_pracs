```python
import cv2
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("/content/Bandra-Worli-Sea-Link-A-masterpiece-of-innovative-civil-engineering-Digpu-News.jpg")
```

```python
plt.imshow(image)
```

```python
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
gray_image=cv2.cvtColor(image_rgb,cv2.COLOR_RGB2GRAY)
plt.imshow(gray_image, cmap="gray")
```

```python
crop = image[30:500,100:900]
plt.imshow(crop)
```

```python
taj=cv2.imread("taj mahal.jpg")
plt.imshow(taj)
```

```python
taj=cv2.cvtColor(taj,cv2.COLOR_BGR2RGB)
plt.imshow(taj)
```

Checking shape of image

```python
taj.shape
```

```python
image.shape
```

```python
taj = cv2.resize(taj,(1200,675))
taj.shape
```

Weighted Sum

```python
weightedSum = cv2.addWeighted(taj,0.5, image_rgb,0.4,0)
plt.imshow(weightedSum)
```

Arithmetic Operations

```python
add=cv2.add(taj,image_rgb)
plt.imshow(add)
```

```python
sub = cv2.subtract(taj, image_rgb)
plt.imshow(sub)
```

Logical Operations

```python
bitwise_and=cv2.bitwise_and(taj,image_rgb)
plt.imshow(bitwise_and)
```

```python
bitwise_or=cv2.bitwise_or(taj,image_rgb)
plt.imshow(bitwise_or)
```

```python
bitwise_xor=cv2.bitwise_xor(taj,image_rgb)
plt.imshow(bitwise_xor)
```

```python
bitwise_not=cv2.bitwise_not(image_rgb)
plt.imshow(bitwise_not)
```

```python
bitwise_not=cv2.bitwise_not(taj)
plt.imshow(bitwise_not)
```
