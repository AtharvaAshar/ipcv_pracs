Contrast Stretching 

```python
def pixelVal(pic,a,b,l,m,n):
  v=l*a
  w=m*(b-a)+v
  if (0<= pic and pic <= a):
      return (v / a)*pic
  elif (a < pic and pic <= b):
      return ((w - v)/(b - a)) * (pic - a) + v
  else:
      return ((255 - w)/(255 - b)) * (pic - b) + w
      
```

```python
a = 60
b = 140 
l_1 = 0.6
m_1 = 1.5 
n_1 = 0.6
image2.min()
```

```python
pixelVal_vec = np.vectorize(pixelVal)
contrast_stretched = pixelVal_vec(image2, a, b, l_1, m_1, n_1)
plt.imshow(contrast_stretched, cmap='gray')
```
